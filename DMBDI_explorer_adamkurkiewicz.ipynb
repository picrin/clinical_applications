```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def probe_expression_level(experiment, gene_name):
    dmpk = pandas.read_csv("analysis/{experiment}/{gene_name}".format(experiment=experiment, gene_name=gene_name), sep="\t")
    just_data = dmpk.filter(regex=("patient*"))
    just_metadata = dmpk.filter(regex=("^((?!patient).)*$"))
    just_metadata = just_metadata.filter(regex="^((?!Unnamed).)*$")
    x = just_metadata.filter(regex=("genocode_left"))
    y = just_data.sum(1)/just_data.shape[1]
    return x, y
```

```python
def scatter_expression_level(gene_name):
    plt.figure()
    plt.scatter(*probe_expression_level("experiment_muscle", gene_name), color="red", label="muscle")
    plt.scatter(*probe_expression_level("experiment_blood", gene_name), color="blue", label="blood")
    plt.legend()
```
